```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
pdf_path = "../data/software_notes.pdf"  # loading our pdf 

loader = PyPDFLoader(pdf_path)
documents = loader.load()

print(f"Total pages loaded: {len(documents)}")
print("Sample extracted text:\n")
print(documents[0].page_content[:800])   # this will give the first page text of the pdf, if it successfully loads the pdf.
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=100,
    separators=["\n\n", "\n", ".", " "],
)

chunks = text_splitter.split_documents(documents)

print(f"Total chunks created: {len(chunks)}")
processed_chunks = [
    chunk.page_content.replace("\n", " ").strip() # This formats the output and present it in a much better way otherwise the paragraph can break anywhere. 
    for chunk in chunks
]

print(processed_chunks[0][:800])
```
