```python
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_community.vectorstores import FAISS #FAISS = vector store
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")

vectorstore = FAISS.load_local(
    "../vector_store/faiss_store",
    embeddings,
    allow_dangerous_deserialization=True
)
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)
"""Here,temperature is used to control our model's creativity of writing the answers
   For ex :
    0-0.4 -> logical & factual
    0.4-0.7 -> story & creative
    0.8-0.12 -> more creative , but can be inaccurate
"""
```

```python
def summarize_query(query, k=4):

    docs = vectorstore.similarity_search(query, k=k) #fetches the most relevant chunks from the vector store
    combined_text = "\n\n".join([d.page_content for d in docs])

    prompt = f""" 
    You are an AI study assistant.
    Summarize the following content in clean bullet points and explain clearly:

    Content:
    {combined_text}

    Summary:
    """
    response = llm.invoke(prompt) #Summarizing everything related the question using the llm
    return response.content
```

```python
print(summarize_query("Summarize the document"))
print(summarize_query("List important concepts"))
print(summarize_query("Explain SDLC from the PDF"))
print(summarize_query("Extract definitions and key terms"))
```
